Cleaning gender data by assigning only F or M character to bring consistency.

```python
def clean_gender_value(gender):
    
    if gender in ["M", "Male", "m"]:
        return "M"
    elif gender in ["F", "Female", "female", "f", "femal", "Femal"]:
        return "F"
    else:
        return "U"
    
    
data["gender"] = list(map(clean_gender_value, data["gender"]))
data["gender"].value_counts()
```
